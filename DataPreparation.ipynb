```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data/final.csv")
data.head()
```

```python
# drop ID
data.drop(["ID"], axis=1, inplace=True)
data.head()
```

```python
data.dropna(inplace=True, ignore_index=True)
```

```python
# check data information
data.info()
```

**Need to encode non-numeric features: Property type, address, law doc, city**

*(?)split post date to get year and quarter, or use datetime data type*

```python
post_dates = data["Post Date"].to_list()
print(len(post_dates))
```

```python
year = []
quarter = []
for date in post_dates:
    d = date.split("-")
    year.append(int(d[0]))
    month = int(d[1])
    if 1 <= month <= 3: quarter.append(1)
    elif 4 <= month <= 6: quarter.append(2)
    elif 7 <= month <= 9: quarter.append(3)
    elif 10 <= month <= 12: quarter.append(4)
    else: quarter.append(0)

year = pd.Series(year, dtype=np.float64)
quarter = pd.Series(quarter, dtype=np.float64)

data.insert(loc=7, column="Year", value=year)
data.insert(loc=7, column="Quarter", value=quarter)
data.drop(["Post Date"], axis=1, inplace=True)

data.head()
```

```python
data["Address"].unique().shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1, encoded_missing_value=-1)
```

```python
encoded_features = encoder.fit_transform(data[["Property Type", "Address", "Law Document", "City"]])
print(encoded_features)
```

```python
encoder.categories_
```

```python
data["Property Type"] = encoded_features[:, 0]
data["Address"] = encoded_features[:, 1]
data["Law Document"] = encoded_features[:, 2]
data["City"] = encoded_features[:, 3]
```

```python
data.info()
```

```python
# save encoder and data
import pickle as pkl
```

```python
pkl.dump(encoder, open("CategoricalDataEncoder.h5", 'wb'))
```

```python
X = data.iloc[:, 1:]
y = data.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train.insert(loc=X_train.shape[1], column="Price (billion VND)", value=y_train)
X_test.insert(loc=X_test.shape[1], column="Price (billion VND)", value=y_test)

X_train.to_csv("data/train_data.csv")
X_test.to_csv("data/test_data.csv")
```
